```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('c:\\Users\\user\\Downloads\\Used_Bikes.csv')
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['brand'].value_counts().keys()
```

```python
brand_list=['Bajaj', 'Royal Enfield', 'Hero', 'Honda', 'Yamaha', 'TVS', 'KTM',
       'Suzuki', 'Harley-Davidson', 'Kawasaki', 'Hyosung', 'Mahindra',
       'Benelli', 'Triumph', 'Ducati', 'BMW']
brand_list
```

```python
df=df[df["brand"].isin(brand_list)]
df
```

```python
categorial=df.select_dtypes(include="O")    # catgorical data types 
categorial
```

```python
numerical=df.select_dtypes(exclude="O")    # numerical data types
numerical
```

```python
categorial['owner'].value_counts()
```

```python
categorial['brand']
```

```python
dt1={"First Owner":1,"Second Owner":2,"Third Owner":3,"Fourth Owner Or More":4}
categorial['owner']= categorial['owner'].map(dt1)
```

```python
categorial
```

```python
categorial["brand"].value_counts().keys()
```

```python
brand_list
dt={}
for i in range(0,len(brand_list)):
    dt[brand_list[i]]=i
```

```python
dt
```

```python
categorial['brand'] = categorial['brand'].map(dt)
categorial
```

```python
categorial['Owner_name']="sanjana"
categorial
```

```python
categorial.drop(['bike_name','city','Owner_name'],axis=1,inplace=True)
categorial
```

```python
data=pd.concat([categorial,numerical],axis=1)
data
```

```python
x = data.drop('price' , axis= 1)
```

```python
y = data[['price']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# x and y --> training and tasting split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)   #test size -> 20 % and other reamaining part - trainning size = 80%
#preduction -> by head   for example when we pass x as arugement get by head y by machine learning algo# x, y for testing and for trainnig only y head(actual value)
# x, y for testing and for trainnig only y head(actual value)
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lg = LinearRegression()
```

```python
x_train.isnull().sum()
```

```python
# Algorithym training
lg.fit(x_train , y_train)    # training data for training purpose
```

```python
# model Testing step -> model evaluation step:
# after complete training of model we start evalautions of our model for check it's performance and efficency
# model evaluation depends on different technique algo
    # Regression model
```

```python
# Training and Testing
print("Training score is:" , round(lg.score(x_train , y_train)*100 ,2))
print("Testing score is:" , round(lg.score(x_test, y_test)*100 ,2))

## you need grater then 90%
# accuracy > 90%
```

```python
y_hat = lg.predict(x_test)   
```

```python
x_test.shape
```

```python
y_test["preduction"] = y_hat
y_test
```

```python
y_tester = y_test['preduction'].map(lambda x: round(x,2))
y_tester
```

```python
y_tester.head()
```

```python
print("Training score is:" , round(rdf.score(x_train , y_train)*100 ,2))
print("Testing score is:" , round(rdf.score(x_test, y_test['price'])*100 ,2))
```
